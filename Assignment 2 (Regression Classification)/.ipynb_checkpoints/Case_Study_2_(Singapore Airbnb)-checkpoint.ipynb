# Case Study 2 ("Regression/Analysis Singapore Airbnb") 

#### 1. Import modules needed

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

#### 2. Load data from 'listings.csv'

```python
sgData=pd.read_csv("listings.csv")
```

#### 3. Print all columns names available from the data

```python
print(sgData.columns)
```

#### 4. Select only columns whose datatype is numeric

```python
newSgData = sgData.select_dtypes(include=[np.number])
```

#### 5. Drop columns that are not needed (id and host id) for the regression. Also Find how many nulls exist in our data

```python
print(newSgData.isna().sum().sum())
print(newSgData.isna().sum())
newSgData=newSgData.drop(columns=["id", "host_id"])
print(newSgData.isna().sum())
```

#### 6. Since the number of null data is more than 10%, we fill the null data with zero

```python
newSgData1 = newSgData.fillna(0)
print(newSgData1.isna().values.any())
print(newSgData1.isna().sum())
```

#### 7. Visualize each column in a histogram 

```python
newSgData1.hist()
plt.subplots_adjust(hspace = 0.8, wspace = 0.8)
plt.rcParams["figure.figsize"] = [16,9]
plt.show()
```

#### 8. Visualize Choosen Data (Price and Number of Reviews) in Scatter

```python
plt.subplot(1,1,1)
plt.scatter(newSgData1["price"], newSgData1["minimum_nights"], color="blue")
plt.xlabel("Price")
plt.ylabel("Minimum Nights")
plt.show()
```

#### 9. Do the regression and also find coefficient and intercept

```python
train, test = train_test_split(newSgData1, test_size=0.2)
regression = linear_model.LinearRegression()
regression.fit(train[["price"]], train[["minimum_nights"]])
print('Coefficients: ', regression.coef_)
print('Intercept: ',regression.intercept_)
```

#### 10. Plot the linear regression result

```python
plt.scatter(train["price"], train["minimum_nights"],  color='blue')
plt.plot(train[["price"]], regression.coef_ * train[["minimum_nights"]] + regression.intercept_ , '-r')
plt.xlim(200, 400)
plt.ylim(0, 300)
plt.xlabel("Price")
plt.ylabel("Minimum Nights")
plt.show()
```

#### 11. Predict for given value and find mean absolute error, mean squared error, r2

```python
prediction = regression.predict(newSgData1[["price"]])
for i in range(len(newSgData1)):
  print(newSgData1[["price"]].values[i], prediction[i])

print("Mean Absolute Error : ", mean_absolute_error(newSgData[["minimum_nights"]], prediction))
print("Mean Squared Error : ", mean_squared_error(newSgData[["minimum_nights"]], prediction))
print("R2 : ", r2_score(newSgData[["minimum_nights"]], prediction))
```
