### Univariate using SKLearn

```python
x = insuranceData[["smoker"]]
y =insuranceData[["charges"]]
xTrain, xTest, yTrain, yTest=train_test_split(x,y, test_size=0.2)
```

```python
regression= LinearRegression().fit(xTrain,yTrain)
```

```python
testing = regression.predict(xTest)
print("Y intercept: ",regression.intercept_)
print("Slope: ", regression.coef_)
print("R2 Score: ", r2_score(yTest,testing))
print("Error: ", mean_squared_error(yTest,testing))
```
